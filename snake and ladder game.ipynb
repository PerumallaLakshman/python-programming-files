# create a python program to play a snake and ladder game between 2 players.

```python
import random
snake_position={ 28:10,
                   39:17,
                   50:5,
                   63:18,
                   74:25,
                   84:58,
                   87:49,
                   92:74,
                   95:56,
                   98:40 }

ladder_position={2:23,
                6:36,
                9:30,
                20:59,
                34:72,
                52:85,
                61:99,
                71:92,
                82:96}

p1_current_position = 0
p2_current_position = 0

p1=input("enter player1 name: ")
p2=input("enter player2 name: ")
while True:
    input("\n"+ p1 +" "+ "press enter to roll the die")
    dice_value=random.randint(1,6)
    print("\n","dice value is",dice_value)
    old_position= p1_current_position
    p1_current_position = p1_current_position + dice_value  
    
    if (p1_current_position > 100):
        print("you need",100-p1_current_position)
        p1_current_position = old_position
        
    if p1_current_position in snake_position:
        p1_current_position=snake_position[p1_current_position]
        print("~~SNAKE BITE~~")
        
    if p1_current_position in ladder_position:
        p1_current_position=ladder_position[p1_current_position]
        print("#######LADDER CLIMB#######")
        
    print(p1,"current position is",p1_current_position)
        
    if (p1_current_position == 100):
        print("congratulations!", p1 ,"has won the game")
        break
        
    input("\n"+ p2 +" "+ "press enter to roll the die")
    dice_value=random.randint(1,6)
    print("\n","dice value is",dice_value)
    old_position= p2_current_position
    p2_current_position = p2_current_position + dice_value
    
    if (p2_current_position > 100):
        print("you need",100-p2_current_position)
        p2_current_position = old_position
        
        
    if p2_current_position in snake_position:
        p2_current_position=snake_position[p2_current_position]
        print("~~SNAKE BITE~~")
        
    if p2_current_position in ladder_position:
        p2_current_position=ladder_position[p2_current_position]
        print("#######LADDER CLIMB#######")
              
    print(p2,"current position is",p2_current_position)
    
    if (p2_current_position == 100):
        print("congratulations!", p2 ,"has won the game")
        break
    p1_current_position = p1_current_position + dice_value  
    
    if (p1_current_position > 100):
        print("you need",100-p1_current_position)
        p1_current_position = old_position
        
    if p1_current_position in snake_position:
        p1_current_position=snake_position[p1_current_position]
        print("~~SNAKE BITE~~")
        
    if p1_current_position in ladder_position:
        p1_current_position=ladder_position[p1_current_position]
        print("#######LADDER CLIMB#######")
        
    print(p1,"current position is",p1_current_position)
        
    if (p1_current_position == 100):
        print("congratulations!", p1 ,"has won the game")
        break
        
    input("\n"+ p2 +" "+ "press enter to roll the die")
    dice_value=random.randint(1,6)
    print("\n","dice value is",dice_value)
    old_position= p2_current_position
    p2_current_position = p2_current_position + dice_value
    
    if (p2_current_position > 100):
        print("you need",100-p2_current_position)
        p2_current_position = old_position
        
        
    if p2_current_position in snake_position:
        p2_current_position=snake_position[p2_current_position]
        print("~~SNAKE BITE~~")
        
    if p2_current_position in ladder_position:
        p2_current_position=ladder_position[p2_current_position]
        print("#######LADDER CLIMB#######")
              
    print(p2,"current position is",p2_current_position)
    
    if (p2_current_position == 100):
        print("congratulations!", p2 ,"has won the game")
        break
```
